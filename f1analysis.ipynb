```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from a CSV file
file_path = '/Users/shashikumarezhil/Downloads/archive 2/circuits.csv'  # Replace with your CSV file path
df = pd.read_csv(file_path)

# Display basic information about the DataFrame
print(df.info())

# Display the first few rows of the DataFrame
print(df.head())

# Display summary statistics for numeric columns
print(df.describe())



# Grouping by country and calculating average elevation
average_altitude = df.groupby('country')['alt'].mean().sort_values()

# Visualizing average elevation by country
plt.figure(figsize=(10, 8))
average_altitude.plot(kind='bar', color='salmon')
plt.ylabel('Average Elevation (meters)')
plt.title('Average Elevation of Racing Circuits by Country')
plt.grid(axis='y')
plt.show()

# Additional analysis: Count the number of circuits by country
circuit_count = df['country'].value_counts()

# Visualizing the number of circuits per country
plt.figure(figsize=(10, 6))
circuit_count.plot(kind='bar', color='lightgreen')
plt.ylabel('Number of Circuits')
plt.title('Number of Racing Circuits by Country')
plt.grid(axis='y')
plt.show()
```

```python
# Visualize the elevation (alt) of circuits
plt.figure(figsize=(20, 26))
plt.barh(df['name'], df['alt'], color='skyblue')
plt.xlabel('Elevation (meters)')
plt.title('Elevation of Various Racing Circuits')
plt.grid(axis='x')
plt.show()
```

```python
# Visualizing the locations of the circuits
plt.figure(figsize=(12, 8))
plt.scatter(df['lng'], df['lat'], color='purple', alpha=0.6, edgecolors='w', s=100)
plt.title('Location of Racing Circuits')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid()
plt.show()
```
